```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**LEER ARCHIVO DE VARIABLES**

```python
#Carga desde un archivo .xlsx sin índice
df= pd.read_csv('listings.csv')
df
```

**A.- Los Registros de 5 "host name" diferentes.**

```python
filtro1= df[(df["host_name"]== "France") | (df["host_name"]== "Marc") | (df["host_name"]=="Mary") | (df["host_name"]=="Judy") | (df["host_name"]=="Honore")]
filtro1
```

**B.- Los registros de los host que se hallan unido a Airbnb "antes del año 2019"**

```python

filtro3= df.iloc[:, [12]]
filtro3
```

```python
df["host_since"] = pd.to_datetime(df["host_since"])
filtro2= df[df["host_since"] <= "01/01/2019"]
filtro2
```

**C.- Los registros de los host que responden "a más tardar en 1 día"**

```python
filtro5= df.iloc[75:100, [15]]
filtro5
```

```python
filtro4= df[(df["host_response_time"] == "within an hour") | (df["host_response_time"] == "within a few hours")]
filtro4
```

**D.- Los registros que tengan "al menos 5 accommodates"**

```python
filtro5=df[df["accommodates"] >= 5]
filtro5
```

**E.- Los registros de los tipos de cuartos: "Entire home" y "Private room"**

```python
filtro6= df[(df["room_type"] == "Entire home/apt") | (df["room_type"]== "Private room")]
filtro6
```

**F.- Los registros que "a lo mucho cuenten con 4 camas"**

```python
filtro7= df[(df["beds"]<=4  )]
filtro7
```

**G.- Los registros que "tengan 1 baño o que sean super_host" pero con la condición de que "todos sean instant_bookable"**

```python
filtro8= df [((df["bathrooms"]==1)|(df["host_is_superhost"]== "t"))&(df["instant_bookable"]== "t")]
filtro8
```

**H.- Los registros cuyo "review_score_cleanlinesssea menor de 3 y review_scores_rating sea mayor a 3"**

```python
filtro9=df[(df["review_scores_cleanliness"] < 3) & (df["review_scores_rating"] > 3)]
filtro9
```
