**ACTIVIDAD 2.1: EXTRACCIÓN DE DATOS**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga desde un archivo .xlsx sin índice
df= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

```python
#Verificamos información del DataFrame
df.info()
```

**FILTROS BÁSICOS**

```python
#Filtro por objeto "LETICIA RAMIREZ HERNANDEZ" de la columna "NOMBRE_VENDEDOR"
filtro1= df[(df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ")]
filtro1
```

```python
#Convertir archivo filtrado a CSV
filtro1.to_csv("filtro1.csv")
```

```python
#Filtrar filas no consecutivas 12,140, 256 y 456.
filtro2=df.iloc[[12,140,256,456], :]
filtro2
```

```python
#Convertir archivo filtrado a CSV
filtro2.to_csv("filtro2.csv")
```

```python
#Filtrar por columnas consecutias de las 2 a las 7
filtro3= df.iloc[ : , 2:7]
filtro3
```

```python
#Convertir archivo filtrado a CSV
filtro3.to_csv("filtro3.csv")
```

```python
#Definir la columna 1 como index.
df1= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col=2)
df1
```

```python
#Filtrar por filas "AARON NUÑO CORDOVA" vs columnas "NOMBRE_CLIENTE"
filtro4= df1.loc[["AARON NUÑO CORDOVA"],["NOMBRE_CLIENTE"]]
filtro4
```

```python
#Convertir archivo filtrado a CSV
filtro4.to_csv("filtro4.csv")
```

```python
#Filtrar por cabecera, es decir las primeras 6 filas.
filtro5=df.head(6)
filtro5
```

```python
#Convertir archivo filtrado a CSV
filtro5.to_csv("filtro5.csv")
```

```python
#Filtrar por cola, es decir las últimas 8 filas.
filtro6=df.tail(8)
filtro6
```

```python
#Convertir archivo filtrado a CSV
filtro6.to_csv("filtro6.csv")
```

```python
#Filtra por comparación, siendo el "PRECIO_UNITARIO" menor a 2000
filtro7= df[df["PRECIO_UNITARIO"] < 2000]
filtro7
```

```python
#Convertir archivo filtrado a CSV
filtro7.to_csv("filtro7.csv")
```

```python
#Filtrar por Y, es decir, filtrar elementos de "COSTO_TOTAL_CALCULADO" menores a 1000 y elementos de "COSTO_UNITARIO_CALCULADO" iguales a 102.31
filtro8=df[(df["COSTO_TOTAL_CALCULADO"] < 1100) & (df["COSTO_UNITARIO_CALCULADO"] == 102.31)]
filtro8
```

```python
#Convertir archivo filtrado a CSV
filtro8.to_csv("filtro8.csv")
```

```python
#Filtro O, es decir, filtrar elementos de "COSTO_TOTAL_CALCULADO" mayores a 2210 o filtrar elementos de "COSTO_UNITARIO_CALCULADO"
filtro9=df[(df["COSTO_TOTAL_CALCULADO"] > 2210) | (df["COSTO_UNITARIO_CALCULADO"] < 1500)]
filtro9
```

```python
#Convertir archivo filtrado a CSV
filtro9.to_csv("filtro9.csv")
```

```python
#Filtro Not, es decir, filtrar elementos de "CANT" diferentes a 6.
filtro10= df[~(df["CANT"] == 6)]
filtro10
```

```python
#Convertir archivo filtrado a CSV
filtro10.to_csv("filtro10.csv")
```
